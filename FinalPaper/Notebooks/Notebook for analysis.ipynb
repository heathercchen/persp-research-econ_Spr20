```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Clean the id dataset
id_df = pd.read_csv('./data/rst_12.csv')
id_df = id_df.sort_values(by=['WAVE'])
id_df
```

```python
id_df.loc[id_df['WAVE'] == 1989]
```

```python
#Seperate id data to individual years dataset
years = ['1989', '1991', '1993', '1997', '2000', '2004', '2006', '2009', '2011', '2015']
for year in years:
    tempid_df = id_df.loc[id_df['WAVE'] == int(year)]
    tempid_df['marital'] = tempid_df['A8']
    tempid_df['rural'] = tempid_df['T2']
    #Only keep individual id, household id, wave no. and spouse id in this dataset
    tempid_df = tempid_df[['IDind', 'WAVE', 'hhid', 'IDind_s', 'marital', 'rural']]
    tempid_df.to_csv('./data/ids/id'+year+'.csv')
```

```python
#Clean the idvidual income dataset
inc_df = pd.read_csv('./data/indinc_10.csv')
inc_df = inc_df.sort_values(by=['wave'])
inc_df
```

```python
for year in years:
    tempinc_df = inc_df.loc[inc_df['wave'] == int(year)]
    #Only keep individual id, household id, wave no. and imputed income (inflated to 2015)
    tempinc_df = tempinc_df[['IDind', 'wave', 'hhid', 'indinc_cpi']]
    tempinc_df.to_csv('./data/incs/inc'+year+'.csv')
```

```python
#Clean the gender dataset
gender_df = pd.read_csv('./data/mast_pub_12.csv')
gender_df = gender_df.sort_values(by=['Idind'])
gender_df = gender_df.rename(columns={'Idind': 'IDind', 'WEST_DOB_Y': 'birth_year', 'GENDER':'gender'})
gender_df = gender_df[['IDind', 'gender', 'birth_year']]
gender_df
```

```python
#Merge birth year and gender data with wave data
for year in years:
    tempinfo_df = pd.read_csv('./data/ids/id'+year+'.csv')
    trial_df = pd.merge(tempinfo_df, gender_df, on='IDind')
    trial_df = trial_df.drop(['Unnamed: 0'], axis=1)
    trial_df.to_csv('./data/infos/info'+year+'.csv')
```

```python
#Clean the job information dataset
job_df = pd.read_csv('./data/jobs_12.csv')
job_df = job_df.sort_values(by=['wave'])
job_df
```

```python
for year in years:
    tempjob_df = job_df.loc[job_df['wave'] == int(year)]
    #Only keep individual id, household id, wave no., current working status (B2) and ever worked (B3)
    tempjob_df = tempjob_df.rename(columns={'B2':'curr_work', 'B3':'ever_work'})
    tempjob_df = tempjob_df[['IDind', 'wave', 'hhid', 'curr_work', 'ever_work']]
    tempjob_df.to_csv('./data/jobs/job'+year+'.csv')
```

```python
#Clean education data
edu_df = pd.read_csv('./data/educ_12.csv')
edu_df = edu_df.sort_values(by=['WAVE'])
edu_df
```

```python
for year in years:
    tempedu_df = edu_df.loc[edu_df['WAVE'] == int(year)]
    #Only keep individual id, household id, wave no., years of schooling and highest degree level
    tempedu_df = tempedu_df.rename(columns={'A11':'schooling', 'A12':'degree'})
    tempedu_df = tempedu_df[['IDind', 'WAVE', 'hhid', 'schooling', 'degree']]
    tempedu_df.to_csv('./data/edus/edu'+year+'.csv')
```

```python
#Clean child data
child_df = pd.read_csv('./data/birthmast_pub_12.csv')
child_df = child_df.sort_values(by=['IDind_m'])
child_df
```

```python
child_df = child_df.rename(columns={'WEST_DOB_Y':'child_birth_year', 's56':'child_death_year'})
child_df = child_df[['IDind_m', 'IDind_c', 'child_birth_year', 'child_death_year']]
child_df
```

### Descriptive statistics
#### 1. Labor market participation of married women

```python
#Merge job dataset with info dataset year by year
for year in years:
    tempinfo_df = pd.read_csv('./data/infos/info'+year+'.csv')
    tempjob_df = pd.read_csv('./data/jobs/job'+year+'.csv')
    tempjob_df = tempjob_df[['IDind', 'curr_work', 'ever_work']]
    temp_df = pd.merge(tempinfo_df, tempjob_df, on='IDind')
    temp_df = temp_df.drop(['Unnamed: 0'], axis=1)
    temp_df.to_csv('./data/emp_info/emp_info'+year+'.csv')
```

```python
mar_df = temp_df.loc[temp_df['marital']==2]
mar_df.loc[mar_df['gender']==2]['curr_work'].value_counts()
```

```python
#Labor participation rate of married men and women
mar_women = []
mar_men = []
for year in years:
    temp_df = pd.read_csv('./data/emp_info/emp_info'+year+'.csv')
    mar_df = temp_df.loc[temp_df['marital']==2]
    mar_women_df = mar_df.loc[mar_df['gender']==2]
    mar_men_df = mar_df.loc[mar_df['gender']==1]
    f_nwork = mar_women_df['curr_work'].value_counts()[0]
    f_work = mar_women_df['curr_work'].value_counts()[1]
    women_rate = f_work / (f_work + f_nwork)
    print('women:', women_rate)
    mar_women.append(women_rate)
    m_nwork = mar_men_df['curr_work'].value_counts()[0]
    m_work = mar_men_df['curr_work'].value_counts()[1]
    men_rate = m_work / (m_work + m_nwork)
    print('men:', men_rate)
    mar_men.append(men_rate)
```

```python
#Plot the graph
plt.plot(years, mar_women, color='red', label='women')
plt.plot(years, mar_men, color='blue', label='men')
plt.xlabel('Years')
plt.ylabel('Labor market participation rate')
plt.legend()
plt.savefig('./graphs/mar_fm.png')
```

```python
never_mar = []
mar = []
div = []
for year in years:
    temp_df = pd.read_csv('./data/emp_info/emp_info'+year+'.csv')
    never_mar_df = temp_df.loc[temp_df['marital']==1]
    mar_df = temp_df.loc[temp_df['marital']==2]
    other_df = temp_df.loc[(temp_df['marital']==3)&(temp_df['marital']==4)&(temp_df['marital']==5)]
    mar_men_df = mar_df.loc[mar_df['gender']==1]
    #Calculate those rates (1.)
    n_nwork = never_mar_df['curr_work'].value_counts()[0]
    n_work = never_mar_df['curr_work'].value_counts()[1]
    never_rate = n_work / (n_work + n_nwork)
    never_mar.append(never_rate)
    #2.
    m_nwork = mar_df['curr_work'].value_counts()[0]
    m_work = mar_df['curr_work'].value_counts()[1]
    mar_rate = m_work / (m_work + m_nwork)
    mar.append(mar_rate)
    #3.
    try:
        d_nwork = other_df['curr_work'].value_counts()[0]
        d_work = other_df['curr_work'].value_counts()[1]
        d_rate = d_work / (d_work + d_nwork)
        div.append(d_rate)
    except KeyError:
        div.append(np.NaN)
```

```python
plt.plot(years, mar, color='green', label='married')
plt.plot(years, never_mar, color='orange', label='never married')
plt.xlabel('Years')
plt.ylabel('Labor market participation rate')
plt.legend()
plt.savefig('./graphs/mar_women.png')
```

```python
import scipy.stats as sts
import scipy.optimize as opt
import scipy.integrate as intgr
```

```python
#Error function of moments
def err_vec(y, x1, x2, x3, x4, b0, b1, b2, b3, b4, simple):
    yhat = b0 + b1*x1 + b2*x2 + b3*x3 + b4*x4
    #print(yhat.shape)
    if simple:
        return y-yhat
    else:
        return (y-yhat)/yhat
    
def criterion(params, *args):
    b0, b1, b2, b3, b4 = params
    y, x1, x2, x3, x4, W = args
    err = err_vec(y, x1, x2, x3, x4, b0, b1, b2, b3, b4, True)
    crit_val = err.T @ W @ err
    return crit_val
```

```python
b0_init, b1_init, b2_init, b3_init, b4_init = 8, 0, 0, 0, 0
params_init = np.array([b0_init, b1_init, b2_init, b3_init, b4_init])
W_hat = np.eye(5346)
gmm_args = (beta_df['ln_wage'], beta_df['exp'], beta_df['exp_sq'], beta_df['schooling_year'], beta_df['rural'], W_hat)
res = opt.minimize(criterion, params_init, args=(gmm_args), tol=1e-14, method='L-BFGS-B')
b0, b1, b2, b3, b4 = res.x
val_crit = res.fun
print("Estimate of b0:",b0)
print("Estimate of b1:",b1)
print("Estimate of b2:",b2)
print("Estimate of b3:",b3)
print("Estimate of b4:",b4)
print('Value of GMM criterion function:',val_crit)
```

```python
beta_df.to_csv('./data/beta2.csv')
beta_df = beta_df.dropna(subset=['curr_work', 'indinc_cpi', 'husband_inc', 'child_under6', 'child_6to18', 'exp', 'schooling_year'], how='all')
id_list = beta_df['IDind'].unique()
total_list2 = []
for id in id_list:
    #print(id)
    id_df = beta_df.loc[beta_df['IDind'] == id]
    id_df['expect_v_work'] = 0
    id_df['expect_v_not'] = 0
    id_df['predict_work'] = 0
    #Sort the dataframe by wave descending
    id_df = id_df.sort_values(by='WAVE', ascending=False)
    waves = id_df['WAVE'].values
    wave_max = id_df['WAVE'].max()
    #print('wave_max:', wave_max)
    id_df_list = []
    for wa_index in range(len(waves)):
        wave = waves[wa_index]
        #print('curr wave:', wave)
        if wave == wave_max:
            curr_df = id_df.loc[id_df['WAVE'] == wave]
            prior_df = id_df.loc[id_df['WAVE'] == waves[wa_index+1]]
            #The predicted value of curr_work:
            value_work = 0.908e15 + (1-3.896e9)*(curr_df['indinc_cpi'] + curr_df['husband_inc']) - 2.154e13*curr_df['exp'] - 3.764e+14*curr_df['child_under6'] - 3.980e14*curr_df['child_6to18'] + 3.730e12*curr_df['schooling_year']
            value_not = curr_df['husband_inc']
            #print("value work:", value_work.values[0])
            #print("value not:", value_not.values[0])
            if value_work.values[0] > value_not.values[0]:
                curr_df.at[curr_df.index, 'predict_work'] = 1
            #Now caculate the expected utility when working 
            #If previous period choose working:
            exp = prior_df['exp'] + 1
            #print(exp)
            v11 = 0.908e15 + (1 - 3.896e9)*(curr_df['indinc_cpi'].values[0] + curr_df['husband_inc'].values[0]) - 2.154e13*exp - 3.764e+14*curr_df['child_under6'].values[0] - 3.980e14*curr_df['child_6to18'].values[0] + 3.730e12*curr_df['schooling_year'].values[0]
            v12 = curr_df['husband_inc']
            utility_work = max(v11.values[0], v12.values[0])
            v21 = 0.908e15 + (1-3.896e9)*(curr_df['indinc_cpi'].values[0] + curr_df['husband_inc'].values[0]) - 2.154e13*prior_df['exp'].values[0] - 3.764e+14*curr_df['child_under6'].values[0]- 3.980e14*curr_df['child_6to18'].values[0] + 3.730e12*curr_df['schooling_year'].values[0]
            v22 = curr_df['husband_inc'].values[0]
            utility_not = max(v21, v22)
            #print('utitlity work', utility_work)
            #print('utitlity not', utility_not)
            curr_df.at[curr_df.index, 'expect_v_work'] = utility_work
            curr_df.at[curr_df.index, 'expect_v_not'] = utility_not
            #print(curr_df)
            id_df_list.append(curr_df)
        elif wa_index < len(waves)-1:
            after_df = id_df_list.pop()
            #print(after_df)
            curr_df = id_df.loc[id_df['WAVE'] == wave]
            prior_df = id_df.loc[id_df['WAVE'] == waves[wa_index+1]]
            #Estimate predicted working status at this period
            crit1 = curr_df['indinc_cpi'] + 0.908e15 - 3.896e9*curr_df['husband_inc'] - 2.154e13*curr_df['exp'] - 3.764e+14*curr_df['child_under6'] - 3.980e14*curr_df['child_6to18'] + 3.730e12*curr_df['schooling_year']
            crit2 = after_df['expect_v_not']-after_df['expect_v_work']
            if crit1.values[0]>crit2.values[0]:
                curr_df.at[curr_df.index, 'predict_work'] = 1
            exp = prior_df['exp'] + 1
            v11 = 0.908e15 + (1 - 3.896e9)*(curr_df['indinc_cpi'].values[0] + curr_df['husband_inc'].values[0]) - 2.154e13*exp - 3.764e+14*curr_df['child_under6'].values[0] - 3.980e14*curr_df['child_6to18'].values[0] + 3.730e12*curr_df['schooling_year'].values[0] + 0.95*after_df['expect_v_work'].values[0]
            v12 = curr_df['husband_inc'].values[0] + after_df['expect_v_not'].values[0]
            utility_work = max(v11.values[0], v12)
            v21 = 0.908e15 + (1-3.896e9)*(curr_df['indinc_cpi'].values[0] + curr_df['husband_inc'].values[0]) - 2.154e13*prior_df['exp'].values[0] - 3.764e+14*curr_df['child_under6'].values[0]- 3.980e14*curr_df['child_6to18'].values[0] + 3.730e12*curr_df['schooling_year'].values[0] + 0.95*after_df['expect_v_work'].values[0]
            v22 = curr_df['husband_inc'].values[0] + after_df['expect_v_not'].values[0]
            utility_not = max(v21, v22)
            curr_df.at[curr_df.index, 'expect_v_work'] = utility_work
            curr_df.at[curr_df.index, 'expect_v_not'] = utility_not
            id_df_list.append(after_df)
            id_df_list.append(curr_df)
            if len(id_df_list) == len(waves)-1:
                total_list2.append(id_df_list)
```
